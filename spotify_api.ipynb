### Fetching Trending Artists Songs

```python
## Function to extract data from JSON and generate table
def generate_data(tracks):
    for item in tracks:
        track_id = item['track']['id']
        track_name = item['track']['name']
        album_name = item['track']['album']['name']
        artists = [artist['name'] for artist in item['track']['artists']]
        popularity = item['track']['popularity']
        release_date = item['track']['album']['release_date']
        duration = item['track']['duration_ms']
        url = item['track']['external_urls']['spotify']

        track_id = []
        track_names = []
        artists_list = []
        album_name = []
        popularity = []
        release_dates = []
        duration = [] 
        url = []

        for item in tracks:
            track_id.append(item['track']['id'])
            track_names.append(item['track']['name'])
            artists_list.append(', '.join([artist['name'] for artist in item['track']['artists']]))
            album_name.append(item['track']['album']['name'])
            popularity.append(item['track']['popularity'])
            release_dates.append(item['track']['album']['release_date'])
            duration.append(item['track']['duration_ms'])
            url.append(item['track']['external_urls']['spotify'])

        df = pd.DataFrame({
            'TrackID' : track_id,
            'TrackName' : track_names,
            'AlbumName' : album_name,
            'ArtistName' : artists_list,
            'ReleaseDate' : release_dates,
            'Duration(ms)': duration,
            'PopularityIndex' : popularity,
            'URL': url
        })
        return df
```

```python
# Function to Generate Tracks Features Table from JSON
def generate_features(features):
    for item in features:
        track_id = item['id']
        danceability = item['danceability']
        energy = item['energy']
        key = item['key']
        loudness = item['loudness']
        mode = item['mode']
        speechiness = item['speechiness']
        acousticness = item['acousticness']
        instrumentalness = item['instrumentalness']
        livenss = item['liveness']
        valence = item['valence']
        tempo = item['tempo']
        duration = item['duration_ms']
        time_signature = item['time_signature']

        track_id = []
        danceability = []
        energy = []
        key = []
        loudness = []
        mode = []
        speechiness = []
        acousticness = []
        instrumentalness = []
        livenss = [] 
        valence = []
        tempo = []
        duration = []
        time_signature = []

    for item in features:
        track_id.append(item['id'])
        danceability.append(item['danceability'])
        energy.append(item['energy'])
        key.append(item['key'])
        loudness.append(item['loudness'])
        mode.append(item['mode'])
        speechiness.append(item['speechiness'])
        acousticness.append(item['acousticness'])
        instrumentalness.append(item['instrumentalness'])
        livenss.append( item['liveness'])
        valence.append(item['valence'])
        tempo.append(item['tempo'])
        duration.append(item['duration_ms'])
        time_signature.append(item['time_signature'])


        features_df = pd.DataFrame({
            'id' : track_id,
            'dancebility' : danceability,
            'energy' : energy ,
            'key':key ,
            'loudnes':loudness,
            'mode':mode ,
            'speech':speechiness ,
            'acoustic':acousticness ,
            'instrument':instrumentalness ,
            'livenss':livenss  ,
            'valance':valence ,
            'tempo':tempo,
            'duration': duration,
            'time_signature': time_signature
        })
        
    return features_df
```
